```python
import pandas as pd
import scipy.stats as st
import numpy as np
import hvplot.pandas
import geoviews as gv
import warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
path = "CSV/zip_codes.csv"
```

```python
zipcode_df = pd.read_csv(path)
```

```python
pd.options.display.max_columns = 700
zipcode_df.head()
```

```python
newcol = []
for i in range(len(zipcode_df["Date"])):
    newcol.append(int(zipcode_df["Date"][i].split("/")[2]))
zipcode_df["Year"] = newcol
zipcode_df

data_df = zipcode_df.loc[zipcode_df["Year"].isin([2018, 2019, 2020, 2021, 2022])]
data_df.head()
```

# FIVE YEAR DATA

```python
#Create smaller dataframe for plotting
five_df = data_df.loc[:, ["Crash Record ID", "Make", "Model", "Severity", "Damages", "Primary Contributory Cause",
                              "Weather Condition", "Lighting Condition",
                              "Injury Type", "Latitude", "Longitude", "Zip Code", "Time", "Year",
                             "Posted Speed Limit"]]

reduced_five = five_df.drop_duplicates(subset=["Crash Record ID"])
sum_df = reduced_five.loc[reduced_five["Injury Type"].isin(["FATAL"])]
sum_df.head()
```

```python
# Which car was involved in the most fatal collisions over a five year period?
make = sum_df["Make"].mode()
make
```

```python
#What is the most common cause of fatal crashes?
cause = sum_df["Primary Contributory Cause"].mode()
cause
```

```python
# In which year did the most fatal collisions occur? 
year = sum_df["Year"].mode()
year
```

```python
# Which zipcode had the most fatal collisions?
zipc = sum_df["Zip Code"].mode()
zipc
```

```python
five_year_map = sum_df.hvplot.points(
    "Longitude",
     "Latitude",
    geo = True,
    tiles = "EsriStreet",
    frame_width = 800,
    frame_height = 500,
    scale = 0.8,
    size = 80,
    color = "Make",
    hover_cols = ["Model", "Primary Contributing Cause", "Damages", "Time",
                 "Weather Condition", "Lighting Condition", "Zip Code"],
    title = ("Fatal Vehicle Collisions by Car Make (2018-2022)")
)
# Display the map
five_year_map
```

```python
for key in sum_df.keys():
    print(f"key = {key}")
    output = sum_df[key].value_counts().nlargest(10)
    print(f"output = \n{output}\n")
```
