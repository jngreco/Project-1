```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
path1 = "../CSV/Traffic_Crashes_-_Crashes.csv"
path2 = "../CSV/Traffic_Crashes_-_Vehicles.csv"
path3 = "../CSV/Traffic_Crashes_-_People.csv"
```

```python
people_data = pd.read_csv(path3)
pd.set_option('display.max_columns', 50)
people_data.head()
```

```python
crash_data=pd.read_csv(path1)
crash_data.head()
```

```python
people_data.keys()
```

```python
crash_data.keys()
```

```python
genderdat = people_data[['SEX']].copy()
genderdat = genderdat[genderdat.SEX != 'X']
genderdat = genderdat.dropna()
genderdat['Fatalities'] = crash_data['INJURIES_FATAL']
genderdat.head()
```

```python
genderdat['SEX'].value_counts()
```

```python
# Percentage of Crashes by Gender

colors=['orange', 'lightblue'] 
genderdat['SEX'].value_counts().plot(kind='pie', colors=colors,
                                     shadow=True, autopct='%1.1f%%')
plt.title('Total Car Crash Fatalities by Gender  2018-2022')
plt.ylabel('Gender')
plt.show()
```

```python
# Bar Chart for Total Crashes by Gender
genderdat['SEX'].value_counts().plot(kind='bar')
plt.title('Total Car Crash Fatalities by Gender  2018-2022')
plt.xlabel('Gender')
plt.ylabel('Fatalities')
plt.show()
```

```python
# filter dataset for outliers

agedat = people_data[['AGE','INJURY_CLASSIFICATION']].copy()
agedat = agedat[agedat.AGE >= 13]
agedat = agedat[agedat.AGE <= 102]
agedat = agedat.dropna()
agedat.head()
```

```python

plt.figure(figsize=(20,10))
agedat['AGE'].value_counts().plot(kind='bar')
plt.title('Number of Crashes by Age 2018-2022')
plt.xlabel('Age')
plt.ylabel('Number of Auto Crashes')
plt.show()
```
